Jupyter notebook for analyzing the jobs in a country for a particular jobtype. Dice.com provides a convinient REST API to retrieve jobs in json format. The jobs retrieved from Dice API are cleaned and stored in MongoDB as json documents. MongoDB's aggregation framework is used to do the analysis. Python 3.6 along with Pymongo driver is used to interact with MongoDB. Finally, plots are displayed using Matplotlib and seaborn packages.

Steps to run the analysis for a particular jobtype (eg: “Data Scientist”) in US. At this time only US is supported.

Assumptions:

Mongodb is installed and running on port default port 27017
Python 3 is installed and pymongo 3.4 is used (other 3.+ versions should also work fine)
Matplotlib and seaborn packages are installed. (should already be installed if you are using anaconda python 3)

Step 1: Launch a new session of Jupyter Notebook or Python terminal (if using python terminal, the bar charts may not be displayed). Alternatively, you can use any IDE like pyCharm/Rodeo.

Step 2: Import the analysis classes

```python
from analysisCountry import AnalysisCountry
```

Step 3: Create the desired analysis object by giving the jobtype and country. Next invoke retrieveAndStoreJobs to get jobs from Dice API and store them into MongoDB.
NOTE: Inserting ~30,000 documents takes about 10mins. So please wait until you get the count of the jobs inserted.

```python
# Create country-level analysis object and pass the jobtype and country code
country = AnalysisCountry("Data Scientist", "US")

# retrieve jobs from the api and store them in mongodb
country.retrieveAndStoreJobs()
```

The country.retrieveAndStoreJobs() call, retrieved jobs from the dice API using the above mentioned URL. It then drops the existing collection, if any, to start fresh. After the jobs are successfully inserted into MongoDB's jobs database -> datascientistUS collection, the count of the number of jobs inserted is displayed.

Step 4: After the jobs are stored into MongoDB, analysis can be performed on the data. Using MongoDB's aggregation framework, the jobs are grouped based on state and the top company within every state who are hiring data scientist is displayed.

```python
# get top company by state
country.getTopCompanyByState()
```

Step 5: If we want to find out the total number of jobs per state within US, that have `Data Scientist` requirement, we can run the following:

```python
# get number of jobs per state
country.getJobsByState(barplot=False)
country.getJobsByState()
```
